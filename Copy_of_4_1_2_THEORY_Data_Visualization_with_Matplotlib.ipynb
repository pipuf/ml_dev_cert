```python
fig = plt.figure()
# large axes
axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)

# small axes (the number is in relation of the axes1)
# for example: 0.2 is 20% of 1
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])

plt.show()
```

```python
fig = plt.figure()
# large axes
axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)
# Small axes
axes2 = fig.add_axes([0.2,0.5,0.25,0.25])

plt.show()
```

```python
fig = plt.figure()
# The bigger Plot
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
# The smaller one
axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(x,y)

plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
# Changing the % of relation
axes2 = fig.add_axes([0.2,0.2,0.5,0.25])
axes2.plot(x,y)

plt.show()
```

```python
fig = plt.figure()
# large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
# Extra tools
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('Years')
axes1.set_ylabel('Salary')
axes1.set_title('Some Tittle')

axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(x,y)

plt.show()
```

```python
fig = plt.figure()
# large axes
axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('Years')
axes1.set_ylabel('Salary')
axes1.set_title('Some Tittle')

axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(a,b)
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed in 1 to 2')

plt.show()
```

```python
fig = plt.figure()
# large axes
axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('Years')
axes1.set_ylabel('Salary')
axes1.set_title('Some Tittle')

axes2 = fig.add_axes([1,1,2,2])
axes2.plot(a,b)
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('Years')
axes2.set_ylabel('Salary')
axes2.set_title('Zoomed in 1 to 2')

plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)
```

```python
# Add some quality/definition
fig = plt.figure(dpi=100)

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)
```

```python
# Also you can change the Size
fig = plt.figure(figsize=(1,8),dpi=200)

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)
```

```python
fig = plt.figure(figsize=(10,3),dpi=100)

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)
```

```python
fig = plt.figure(figsize=(2,2),dpi=100)

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)
# To save it
fig.savefig('new_fig.png')
```

```python
fig = plt.figure(figsize=(10,3),dpi=100)

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)
# bbox_inches = 'tight' FOR SHOWING THE AXES INFORMATION IN PNG
fig.savefig('new_fig1.png',bbox_inches='tight')
```
