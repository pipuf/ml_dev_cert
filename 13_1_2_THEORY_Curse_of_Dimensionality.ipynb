# What is the curse of dimensionality?

The curse of dimensionality refers to phenomena that occur when classifying, organizing and analyzing `high dimensional` data that do not occur in `low dimensional` spaces, specifically the issue of `data sparsity` and `data certainty.

# Issues
Data sparsity occurs when moving to higher dimensions. The `volume of the space represented grows so fast` that the data cannot be maintained and therefore becomes sparse. The sparsity problem is important for any problem whose objective has some statistical significance.

As a data space moves from one dimension to two dimensions and then to three dimensions, the given data occupy less and less data space. To maintain an accurate representation of the space, the data for analysis grows exponentially.

The second problem that arises is related to data classification. In low-dimensional spaces, the data can look very similar, but the higher the dimension, the farther apart these data points can appear. The two wind turbines below look very close to each other in two dimensions, but far apart when viewed in a third dimension. This is the same effect that the curse of dimensionality has on the data.

# Infinite features require infinite training

When creating neural networks, instances are created with a certain number of features (dimensions). Each feature has individual aspects, each aspect falling somewhere along each dimension. In a fruit classification example, we may want one feature to handle color, one for weight, one for shape, etc. Each characteristic adds information, and if we could handle all possible characteristics, we could know perfectly well which fruit we are considering. However, an infinite number of features requires an infinite number of training examples, which eliminates the real-world utility of our network.

Most disconcerting is that the amount of training data needed increases exponentially with each added feature. Even if we only had 15 features, each of which is a "yes" or "no" question about the fruit we are identifying.

# Mitigating the Curse of Dimensionality

A careful choice of the number of dimensions (features) to be used is the prerogative of the data scientist training a network. In general, the smaller the size of the training set, the fewer features you should use. You should keep in mind that each feature increases the requirement of the data set exponentially.

# Python Example

```python
import pandas as pd
import numpy as np
%matplotlib inline
#calling it a second time may prevent some graphics errors
%matplotlib inline
import matplotlib.pyplot as plt
from numpy.random import seed
from numpy.random import randn
from mpl_toolkits.mplot3d import Axes3D
seed(42)
```

```python
# First, I generate the DATA
i = 100 # Amount of DATA
sigma = 10
mu = 0
X = randn(i)*sigma+mu
y = randn(i)*sigma+mu
z = randn(i)*sigma+mu
```

# 1 Dimension

```python
plt.figure(figsize=(10,5))
plt.scatter(X, [0]*i);
plt.show()
```

# 2 Dimensions

```python
plt.figure(figsize=(10,5))
plt.scatter(X, y);
plt.show()
```

# 3 Dimensions

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X, y, z,s=60);
plt.show()
```
